```python
from autoscraper import AutoScraper
```

```python
amazon_url = "https://www.amazon.in/s?k=iphone"
wanted_list = ['₹59,999',
               'Apple iPhone 12 (128GB) - Blue',
               '9,256']
```

```python
scraper = AutoScraper()
result = scraper.build(url= amazon_url, wanted_list=wanted_list)
result
```

```python
scraper.get_result_similar(amazon_url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_gjcq':'Title', 'rule_zoue':'Price', 'rule_6uph':'Ratings'})
scraper.keep_rules(['rule_gjcq', 'rule_zoue', 'rule_6uph'])
scraper.save('amazon-search')
```

```python
results = scraper.get_result_similar(amazon_url, group_by_alias=True)
```
